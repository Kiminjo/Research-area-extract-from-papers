```python
import pandas as pd
```

```python
nodes = pd.read_csv('network/nodes_table.csv')
```

```python
raw_data = pd.read_csv('data/2019-2020.csv')
```

```python
def get_abstract(df) :
    abstract = []
    for idx in df['Id'] :       
        abstract.append(raw_data['Abstract'].loc[int(idx)])
    return abstract
```

```python
def doc_count(comm, word) :
    count = 0
    for doc in comm :
        if doc.count(word) >= 3 :
            count += 1
    return count
```

```python
import nltk
from nltk.corpus import stopwords

def get_words(data) :
    clean_data = []
    tokens = []
    for doc in data :
        tokens.append(nltk.regexp_tokenize(doc.lower(), '[A-Za-z]+'))
    for words in tokens :
        clean_words = []
        for word in words :
            if len(word) > 1 :
                if word not in stopwords.words('english'):
                    clean_words.append(word)
        clean_data.append(clean_words)
    return clean_data
```

```python
keywords_df = pd.DataFrame(columns = ['modularity_class', 'proportions', 'keywords'])
```

```python
n_comm = 34

for i in range(n_comm) :
    comm = nodes.loc[nodes['modularity_class'] == i]
    abst = get_words(get_abstract(comm))
    keywords_c = []
    
    for doc in abst :
        for word in doc :
            tmp = []
            cnt = doc_count(abst, word)
            if cnt > 0 :
                tmp.append(word)
                tmp.append(cnt)                
                if tmp not in keywords_c :
                    keywords_c.append(tmp)
                                
    keywords_c.sort(key=lambda x:x[1], reverse=True)
    
    print('\nKeywords of community ' + str(i) + ' : ')
    if len(comm) < 3 :
        print('Warning! The number of document is less than 3.')
        if len(comm) == 1 :
            print('The number of document is only 1.')
    
    prev = 0
    rank = 0
    count = 0
    max_rank = 5
    words = ''
    
    for word in keywords_c :
        count += 1
        if count > 20 :
            max_rank -= 1
            if count > 30 :
                max_rank -= 1
        if prev != word[1] :
            rank += 1
        if rank <= max_rank :
            print(str(rank) + '. ' + str(word))
            words += str(word[0]) + ', '
        prev = word[1]
    
    prp = float(len(comm) / len(nodes))
    keywords_df.loc[len(keywords_df)] = [i, prp, words]
```
